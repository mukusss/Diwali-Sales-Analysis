```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# change data type#
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#checking of column names#
df.columns
```

```python
#name changing#
df.rename(columns={'Martial_Status': 'Shaadi'})
```

# Gender

```python
sns.countplot(data=df,x='Gender',hue='Gender')
```

```python
ax=sns.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

# from above we can say that female purchase is 4425 more than male)

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Gender')
```

# From above graphs we ca see the most of the buyers are females and even the purchasing power of females are greater than men

# AGE

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount', data = sales_age,hue='Age Group')
```

# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# STATE

```python
#Total number of orders from top 10 states#
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Orders', data = sales_state,hue='State')
```

```python
##Total amount /sales from top 10 states##
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount', data = sales_state,hue='State')
```

# Top 7 states retain its position in both cases, more number of orders as well as the purchasing power.Or we can say that most of the orders and total total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectevily.

# MARITAL STATUS

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status')
for bars in ax.containers:
    sns.set(rc={'figure.figsize':(6,8)})
    ax.bar_label(bars)
```

```python
sales_Marital = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount', data = sales_Marital,hue='Gender')
```

# From above graphs we can see that most of the buyers are married (women)and they have high purchasing power
